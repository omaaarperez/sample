```python
def get_series_score(dataframe):
    series = []
    series_reversed = []
    dataframe.head()
    years = sorted(dataframe['Year'].unique())
    for year in years:
        home_team_score = 0
        away_team_score = 0
        
        games = max(dataframe[dataframe['Year']== year]['Game'])
        
        temp = dataframe[dataframe['Year']==year]
        temp = temp.head(len(temp)//2)
        
        for game in range(games):
            if temp.iloc[game]['Win'] == 1:
                home_team_score += 1
            else: 
                away_team_score += 1
            series_score = "{}-{}".format(home_team_score,away_team_score)
            series_reversed_score = "{}-{}".format(away_team_score,home_team_score)
            series.append(series_score)
            series_reversed.append(series_reversed_score)
            
    return series + series_reversed
```

```python
def get_mean(dataframe):
    mean = []

    dataframe = dataframe[['Year', 'Team', 'Game', 'Home', 'MP', 'FG', 'FGA', 'FGP', 'TP',
                           'TPA', 'TPP', 'FT', 'FTA', 'FTP', 'ORB', 'DRB', 'TRB', 'AST', 'STL',
                           'BLK', 'TOV', 'PF', 'PTS', 'Series']]
    years = sorted(dataframe['Year'].unique())
    for year in years:
        temp = dataframe[dataframe['Year']==year]
        num_of_games = max(temp['Game'])
        
        # Upper - Winning team
        # Lower - Losing Team
        temp_upper = temp.head(len(temp)//2).mean()
        temp_lower =  temp.tail(len(temp)//2).mean()
        
        # Since using mean() will have a wrong value for the game column, we will replace it with the
        # max integer value of the game. Example, If a series has 4 games, using mean would return (1+2+3+4)/4 = 2.5,
        # instead, we will return 4 in this case
        temp_upper[1] = num_of_games 
        temp_lower[1] = num_of_games
        
        upper_home_away = temp['Home'].iloc[0]
        lower_home_away = temp['Home'].iloc[len(temp)//2]
        
        
        # Appending Series result
        series_result = pd.Series(dataframe[dataframe['Year']==year]['Series'].iloc[len(temp)//2-1])
        series_result_flip = pd.Series(dataframe[dataframe['Year']==year]['Series'].iloc[len(temp)-1])
        temp_upper = pd.concat([pd.Series([temp['Team'].iloc[0], upper_home_away, series_result[0]]), temp_upper])
        temp_lower = pd.concat([pd.Series([temp['Team'].iloc[len(temp)-1], lower_home_away, series_result_flip[0]]),
                                temp_lower])
        
        temp_upper = temp_upper.drop(index='Home')
        temp_lower = temp_lower.drop(index='Home')
        
        mean.append(temp_upper)
        mean.append(temp_lower)
    return mean
```
